**2. Print even numbers**

```python
def print_even_number(lst):
    """
    Loop through the given list of numbers and print the even numbers in the original order
    as they appear in the list.
    For example: [6,3,1,4,2] -> [6,4,2]
    """
    for i in lst:
        if (int(i) % 2):
            lst.remove(i)
    return lst


    # your program goes here!
```

**3. Digit sum**

```python
def digit_sum(x):
    """
    This function takes in a positive integer x and returns the sum of
    all the number's digits.
    For example: 257 = 2 + 5 + 7 = 14
    """
    y = str(x)
    digit_sum = 0
    for i in y:
        digit_sum += int(i)
    return digit_sum
    

    # your program goes here!
```

```python
print(digit_sum(257))
```
